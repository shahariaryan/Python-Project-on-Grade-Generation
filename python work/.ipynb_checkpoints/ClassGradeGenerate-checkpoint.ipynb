#assignment marks extraction

```python
adf = pd. read_csv ('Assignment.csv')
sid5=np.array(len(adf.index), dtype='str')
sid5=pd.DataFrame(adf, columns=['Student ID']).to_numpy()
score5=np.array(len(adf.index), dtype=np.int32)
score5=pd.DataFrame(adf, columns=['Ass.']).to_numpy()
name=np.array(len(adf.index), dtype='str')
name=pd.DataFrame(adf, columns=['Name']).to_numpy()
space=" "
sname=" "
for i in range(0,len(adf.index)):
    name[i][0]=name[i][0].split(', ')
    if len(name[i][0])==2:
        sname=np.char.add(name[i][0][1],space)
        name[i]=np.char.add(sname,name[i][0][0])
    else:
        name[i]=name[i][0][0]
```

#Initializing a 2D-Array

#Calculation of Total Marks

#Assigning Grades According to the Total Marks

```python
rows, cols = (len(sid5), 16)
result=[]
for i in range(rows):
    col = []
    for j in range(cols):
        col.append(0)
    result.append(col)

for r in range(0,len(sid5)):
    result[r][0]=sid5[r][0]
    result[r][1]=name[r][0]
    result[r][13]=score5[r][0]
    for q in range(0,len(sid1)):
        if(sid1[q][0]==result[r][0]):
            result[r][8]=score1[q][0]
        else:
            continue
    for q in range(0,len(sid2)):
        if(sid2[q][0]==result[r][0]):
            result[r][9]=score2[q][0]
        else:
            continue
    for q in range(0,len(sid3)):
        if(sid3[q][0]==result[r][0]):
            result[r][10]=score3[q][0]
        else:
            continue
    for q in range(0,len(sid4)):
        if(sid4[q][0]==result[r][0]):
            result[r][12]=score4[q][0]
        else:
            continue
    for q in range(0,len(sid7)):
        if(sid7[q][0]==result[r][0]):
            result[r][2]=score7[q][0]
        else:
            continue
    for q in range(0,len(sid8)):
        if(sid8[q][0]==result[r][0]):
            result[r][3]=score8[q][0]
        else:
            continue
    for q in range(0,len(sid9)):
        if(sid9[q][0]==result[r][0]):
            result[r][4]=score9[q][0]
        else:
            continue
    for q in range(0,len(sid10)):
        if(sid10[q][0]==result[r][0]):
            result[r][5]=score10[q][0]
        else:
            continue
    for q in range(0,len(sid11)):
        if(sid11[q][0]==result[r][0]):
            result[r][6]=score11[q][0]
        else:
            continue
for r in range(0,len(sid5)):
    result[r][7]=(result[r][2]+result[r][3]+result[r][4]+result[r][5]+result[r][6])*2
    list1=[result[r][8],result[r][9],result[r][10]]
    list1.sort()
    result[r][11]=list1[-1]+list1[-2]
    result[r][14]=result[r][13]+result[r][12]+result[r][11]+result[r][7]
    if result[r][14]>=90:
        result[r][15]='A+'
    elif result[r][14]>=85:
        result[r][15]='A'
    elif result[r][14]>=80:
        result[r][15]='B+'
    elif result[r][14]>=75:
        result[r][15]='B'
    elif result[r][14]>=70:
        result[r][15]='C+'
    elif result[r][14]>=65:
        result[r][15]='C'
    elif result[r][14]>=60:
        result[r][15]='D+'
    elif result[r][14]>=50:
        result[r][15]='D'
    else:
        result[r][15]='F'
```

#converting 2D Array into dataframe

#make a excel file of the data of the grade 

```python
df = pd.DataFrame(data=result, columns=["Student ID", "Student Name", "Class 1", "Class 2", 
                                        "Class 3", "Class 4", "Class 5", "Total Attendance", "Quiz1", 
                                        "Quiz2", "Quiz3", "Quiz Total", "Exam", "Assignment", "Total Marks", "Grade"])

df.to_excel (r'GradeSheet.xlsx', index = False, header = True)
```

## Model Design

#Performing label encoding

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
df['Grade']= le.fit_transform(df['Grade'])
df['Student ID']= le.fit_transform(df['Student ID'])
df['Student Name']= le.fit_transform(df['Student ID'])
df
```

#importing few library 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

#Segregating the data into training and testing set

```python
train= df.drop(['Student ID','Student Name'], axis=1)
test=df['Grade']
```

```python
X_train,X_test,y_train,y_test=train_test_split(train,test,test_size=0.3,random_state=2)
```

```python
regr= LinearRegression()
```

```python
regr.fit(X_train,y_train)
```

```python
pred=regr.predict(X_test)
pred
```

#Checking the efficiency of the model 

```python
regr.score(X_test,y_test)
```
